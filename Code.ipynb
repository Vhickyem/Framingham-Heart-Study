## Understand the Dataset and ensure it's clean

```python
# Import all necessary packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Import the file
hearty = pd.read_csv('framingham.csv')
hearty.head(10)
```

```python
# check for shape of the dataset
hearty.shape
```

```python
# check for statistical information on the dataset
hearty.describe()
```

```python
# check for information on the column
hearty.info()
```

```python
# information shows null values are present. How many are there?
hearty.isna().sum()
```

```python
# fill some of the numerical values with their mean value
hearty.fillna({'cigsPerDay' : hearty['cigsPerDay'].mean(),
               'totChol' : hearty['totChol'].mean(),
               'glucose' : hearty['glucose'].mean()},
               inplace = True)
hearty.isna().sum()
```

```python
# drop some null values
hearty.dropna(axis = 0, subset = ['BMI', 'heartRate', 'education', 'BPMeds'], inplace = True)
hearty.isna().sum()
```

```python
hearty.shape
```

Our Dataset is now clean. We'll move on to the next step

## Exploratory Data Analysis

```python
# identify the categorical and numeric variables
cat_var = []
num_var = []
for cols in hearty:
    if len(np.unique(hearty[cols].values)) <= 5:
        cat_var.append(cols)
    else:
        num_var.append(cols)

print('Categorical variables include:\n', cat_var)
print('Numeric variables include:\n', num_var)
```

```python
# check for duplicates
hearty.duplicated().sum()
```

There are no duplicates

```python
# Understand the categorical variables
for item in cat_var:
    sns.countplot(x= item, data= hearty, hue= item, palette= 'summer')
    plt.show()
```

From the Visualizations above, we can observe that:
- The male patients in the survey are fewer than the other gender
- The higher the education level, the lower the number of patients
- There isn't much difference in the numbers of those who smoke and those who don't
- The number of those who are on BP medications is far lower than those who don't
- Those who have a prevalent stroke are fewer than those who don't have
- There are over 1000 people that have prevalent hypertension compared to the less than 500 people who are on BP medications
- There aren't a lot of people with diabetes and very few patients are likely to have the coronary heart disease (CHD) within 10 years

```python
# understand the numerical variables
for item in num_var:
    sns.distplot(hearty[item], kde= False, color='green')
    plt.show()
    print(f"Skewness : {hearty[item].skew()}")
    print(f"Kurtosis : {hearty[item].kurt()}")
```

From the Histograms above, we can observe that 
- Skewness and kurtosis is less than 1 with kurtosis, negative for AGE
- Spikes are present in the amounts of cigarettes taken per day. Skewness is to the right and greater than 1. Kurtosis is almost 1 at 0.99
- Total Cholesterol level is skewed to the right. Skewness is less than 1 and Kurtosis is higher than 1.
- SysBP is skewed to the right. Skeness and Kurtosis values are both positive and greater than 1.
- HeartRate has spikes. Kurtosis and Skewness are less than 1.
- Glucose is a bit unusual as it has its skewness greater than 6 and Kurtosis, higher than 66. It also has a spike.

```python
# Study relationships between variables
sns.pairplot(hearty)
plt.show()
```

```python
for num in num_var:
    sns.catplot(x= 'TenYearCHD', y= num, data= hearty, kind= 'box')
    plt.show()
```

From the above, it can be inferred that, Those patients with a risk of having Coronary heart disease within ten years 
- are Older in age
- take more cigarettes per day
- have a higher total cholesterol level
- have a higher systolic and diastolic blood pressure
- have a higher median value of BMI
- have a higher median glucose value than those without the risk.

```python
# explore with a heatmap
corrmat = hearty.corr()
hm = sns.heatmap(corrmat,
cbar= True,
annot= True,
fmt= '.2f',
annot_kws= {'size': 7.5},
yticklabels= hearty.columns,
xticklabels= hearty.columns,
cmap= 'RdYlGn')
plt.show()
```

From the above matrix, it makes sense that there is a strong correlation between

## Use Support Vector Machine (SVM) to Predict 10-year risk of Coronary Heart Disease (CHD)

```python
# split dataset
from sklearn.model_selection import train_test_split
y = hearty['TenYearCHD']
x = hearty.drop(['TenYearCHD'], axis= 'columns')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33)
```

```python
print(len(x_train))
print(len(x_test))
```

```python
# Import svm model
from sklearn.svm import SVC
model = SVC()
```

```python
# Train Model
model.fit(x_train, y_train)
```

```python
# Check Model accuracy score
round(model.score(x_test, y_test), 2)
```

Accuracy score of the Model stands at 0.84, even after tuning with regularization and gamma parameters

## Build a Confusion Matrix to see how the model did

```python
from sklearn.metrics import confusion_matrix
y_predicted = model.predict(x_test)

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
sns.heatmap(cm, 
            annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

The model did pretty well, in that, the model predicted, the patients have the risk of having coronary heart disease within ten years, wrongly twice and also predicted the patients don't have the risk, wrongly, just once.
